```python
img = np.ones((50,50,3),dtype='uint8') * 255
img[:25,:25,1:]=0 #= img[:25,:25,1:]*0
img[:25,25:,::2]=0 #= img[:25,25:,::2]*0
img[25:,:25,:2]=0 #= img[:25,:25,1:]*0
img[25:,25:,2:]=0 #= img[:25,:25,1:]*0
plt.imshow(img)
```

```python
plt.imsave('squarebox.png',img)
```

```python
skimage.io.imsave('sqauarebox2.png',img)
```

```python
cv2.imwrite('squarebox3.png',img[:,:,::-1])
```
