```python
#live image capturing
import cv2, skimage as ski
vid = cv2.VideoCapture(0)
while True:
    ack, img = vid.read()
    if ack:
        th, red_bw = cv2.threshold(
            cv2.subtract(
                img[:,:,-1], cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            ), 40 ,255, cv2.THRESH_BINARY
        )
        strel = cv2.getStructuringElement(cv2.MORPH_RECT,(20,20))
        red_bw2 = cv2.morphologyEx(
            red_bw, cv2.MORPH_CLOSE, strel, iterations=1
        )
        red_bw = ski.morphology.remove_small_objects(
            red_bw.astype(bool), 100
        ).astype('uint8')*255
        red_bw = ski.morphology.remove_small_holes(
            red_bw.astype(bool), 100
        ).astype('uint8')*255
        rps = ski.measure.regionprops(
            ski.measure.label(red_bw.astype(bool))
        )
        count = len(rps)
        img_copy =img.copy()
        cv2.putText(img_copy, str(count),(100,100),
                    cv2.FONT_HERSHEY_PLAIN, 5, (0,0,255),10)
        for rp in rps:
            y1,x1,y2,x2 = rp.bbox
            cv2.rectangle(img_copy, (x1,y1),(x2,y2),(255,0,0),thickness=5)

        #do the entire processing code
        #...
        cv2.imshow('Preview', img_copy) #depends on requirement
        key = cv2.waitKey(1)
        if key == ord('x'):
            break
cv2.destroyAllWindows()
vid.release()   
```

```python
# Face Detection - Viola-jones (harcascade features) 
import cv2
fd = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)
vid = cv2.VideoCapture(0)
while True:
    ack, img = vid.read()
    if ack:
        #to tell face detection
        #...
        faces = fd.detectMultiScale(img,1.2,10)
        
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w,y+h),(255,255,255), 10)
        cv2.imshow('preview', img) #depends on requirement
        key = cv2.waitKey(1)
        if key == ord('x'):
            break
cv2.destroyAllWindows()
vid.release() 
```

```python
# Face Detection - Viola-jones (harcascade features) and save
import cv2
fd = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)
vid = cv2.VideoCapture(0)
counter = 0
while True:
    ack, img = vid.read()
    if ack:
        #to tell face detection
        #...
        faces = fd.detectMultiScale(img,1.2,10)
        if len(faces) == 1:
         counter += 1
         for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w,y+h),(255,255,255), 10)
        face_img = img[y:y+h, x:x+w, :].copy()
        cv2.imwrite(
           r'faces/mukesh/face_{0}.png'.format(counter),
           face_img
        )
        if counter == 50:
           break
        cv2.imshow('preview', img) #depends on requirement
        key = cv2.waitKey(1)
        if key == ord('x'):
            break
cv2.destroyAllWindows()
vid.release() 
```
