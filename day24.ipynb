```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = 'https://raw.githubusercontent.com/redashu/Datasets/master/advertising.csv'
df = pd.read_csv(dataset)
```

```python
df
```

```python
labels=df['Sales']
feature=df[['TV','Newspaper','Radio']]
```

```python
train_X,test_x,train_Y,test_y=train_test_split(feature,labels,test_size=0.2,random_state=2)
```

```python
model=LinearRegression()
```

```python
train_model=model.fit(train_X,train_Y)
```

```python
print(dir(train_model))
```

```python
a=train_model.predict(test_x)
```

```python
#for cofficent

coefficients = train_model.coef_ #m
intercept = train_model.intercept_   #c
```

```python
plt.scatter(feature['TV'], labels)
plt.plot(feature['TV'], coefficients[0] *feature['TV']+ intercept,color='red')
plt.show()
```
